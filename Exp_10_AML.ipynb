```python
!pip install mlxtend


import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx"

df = pd.read_excel(url)

print("========== DATASET HEAD ==========\n")
print(df.head())


print("\n========== DATASET INFO ==========\n")
print(df.info())

print("\n========== MISSING VALUES ==========\n")
print(df.isnull().sum())

print("\n========== DATASET SHAPE ==========\n")
print(df.shape)

# ============================================================
# STEP 5 : DATA CLEANING
# ============================================================

# Remove missing values
df.dropna(inplace=True)

# Remove cancelled invoices
df = df[~df['InvoiceNo'].astype(str).str.contains('C')]

# Remove negative quantities
df = df[df['Quantity'] > 0]

print("\n========== CLEANED DATASET ==========\n")
print(df.head())

# ============================================================
# STEP 6 : EXPLORATORY DATA ANALYSIS (EDA)
# ============================================================

# Top Selling Products

top_products = df['Description'].value_counts().head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top_products.values,
    y=top_products.index
)

plt.title("Top 10 Selling Products")

plt.xlabel("Count")

plt.ylabel("Products")

plt.show()

# Top Countries

top_countries = df['Country'].value_counts().head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    x=top_countries.values,
    y=top_countries.index
)

plt.title("Top Countries by Orders")

plt.xlabel("Orders")

plt.ylabel("Country")

plt.show()

# ============================================================
# STEP 7 : MARKET BASKET CREATION
# ============================================================

# Use only United Kingdom data
basket = (
    df[df['Country'] == 'United Kingdom']

    .groupby(['InvoiceNo', 'Description'])['Quantity']

    .sum()

    .unstack()

    .reset_index()

    .fillna(0)

    .set_index('InvoiceNo')
)

# Convert quantities into binary values

def encode_units(x):

    if x <= 0:
        return 0

    if x >= 1:
        return 1

basket = basket.applymap(encode_units)

print("\n========== BASKET DATA ==========\n")
print(basket.head())

# ============================================================
# STEP 8 : APPLY APRIORI ALGORITHM
# ============================================================

frequent_itemsets = apriori(
    basket,
    min_support=0.02,
    use_colnames=True
)

print("\n========== FREQUENT ITEMSETS ==========\n")
print(frequent_itemsets.head())

# ============================================================
# STEP 9 : GENERATE ASSOCIATION RULES
# ============================================================

rules = association_rules(
    frequent_itemsets,
    metric="lift",
    min_threshold=1
)

# Sort by lift
rules = rules.sort_values(
    by='lift',
    ascending=False
)

print("\n========== ASSOCIATION RULES ==========\n")

print(rules[['antecedents',
             'consequents',
             'support',
             'confidence',
             'lift']].head(10))

# ============================================================
# STEP 10 : VISUALIZE SUPPORT VS CONFIDENCE
# ============================================================

plt.figure(figsize=(8,6))

sns.scatterplot(
    x=rules['support'],
    y=rules['confidence'],
    size=rules['lift'],
    hue=rules['lift']
)

plt.title("Support vs Confidence")

plt.xlabel("Support")

plt.ylabel("Confidence")

plt.show()

# ============================================================
# STEP 11 : INTERPRETATION OF RULES
# ============================================================

print("\n================================================")
print("INTERPRETATION OF ASSOCIATION RULES")
print("================================================\n")

for index, row in rules.head(5).iterrows():

    print("Rule :", set(row['antecedents']),
          "=>",
          set(row['consequents']))

    print("Support :", round(row['support'], 3))

    print("Confidence :", round(row['confidence'], 3))

    print("Lift :", round(row['lift'], 3))

    print()

# ============================================================
# FINAL CONCLUSION
# ============================================================

print("================================================")
print("FINAL CONCLUSION")
print("================================================\n")

print("1. EDA was performed on Online Retail dataset.")
print("2. Apriori algorithm identified frequent itemsets.")
print("3. Association rules revealed product relationships.")
print("4. Rules can help in cross-selling and recommendations.")
print("5. Businesses can optimize promotions and product placement.")
```
