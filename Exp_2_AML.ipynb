```python


import pandas as pd
import numpy as np

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler, LabelEncoder

from sklearn.feature_selection import (
    VarianceThreshold,
    SelectKBest,
    chi2,
    f_classif,
    mutual_info_classif
)



data = load_breast_cancer()


df = pd.DataFrame(data.data, columns=data.feature_names)


df['target'] = data.target

print("========== ORIGINAL DATASET ==========\n")
print(df.head())



# Check Missing Values
print("\n========== MISSING VALUES ==========\n")
print(df.isnull().sum())

# Separate Features and Target
X = df.drop('target', axis=1)
y = df['target']

# Feature Scaling
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("\n========== SCALED FEATURES ==========\n")
print(X_scaled.head())



print("\n================================================")
print("FILTER METHOD 1 : VARIANCE THRESHOLD")
print("================================================\n")

selector_var = VarianceThreshold(threshold=0.5)

selector_var.fit(X_scaled)

selected_features_var = X.columns[selector_var.get_support()]

print("Selected Features:\n")
print(selected_features_var)



print("\n================================================")
print("FILTER METHOD 2 : PEARSON CORRELATION")
print("================================================\n")

correlation = df.corr()['target'].abs().sort_values(ascending=False)

# Remove target itself
top_corr_features = correlation[1:11]

print("Top Correlated Features:\n")
print(top_corr_features)



print("\n================================================")
print("FILTER METHOD 3 : CHI-SQUARE TEST")
print("================================================\n")

# Chi-square requires positive values
X_positive = X.copy()

# Apply SelectKBest with chi2
chi_selector = SelectKBest(score_func=chi2, k=10)

chi_selector.fit(X_positive, y)

chi_scores = pd.DataFrame({
    'Feature': X.columns,
    'Chi2 Score': chi_selector.scores_
})

chi_scores = chi_scores.sort_values(
    by='Chi2 Score',
    ascending=False
)

print("Top Features using Chi-Square:\n")
print(chi_scores.head(10))




print("FILTER METHOD 4 : ANOVA F-TEST")


anova_selector = SelectKBest(score_func=f_classif, k=10)

anova_selector.fit(X_scaled, y)

anova_scores = pd.DataFrame({
    'Feature': X.columns,
    'F-Score': anova_selector.scores_
})

anova_scores = anova_scores.sort_values(
    by='F-Score',
    ascending=False
)

print("Top Features using ANOVA F-Test:\n")
print(anova_scores.head(10))


print("\n================================================")
print("FILTER METHOD 5 : MUTUAL INFORMATION")
print("================================================\n")

mi_scores = mutual_info_classif(X_scaled, y)

mi_df = pd.DataFrame({
    'Feature': X.columns,
    'MI Score': mi_scores
})

mi_df = mi_df.sort_values(
    by='MI Score',
    ascending=False
)

print("Top Features using Mutual Information:\n")
print(mi_df.head(10))



print("\n================================================")
print("FINAL SUMMARY OF FEATURE SELECTION METHODS")
print("================================================\n")

print("1. Variance Threshold Features Count :",
      len(selected_features_var))

print("\n2. Pearson Correlation Top Features :")
print(top_corr_features.index.tolist())

print("\n3. Chi-Square Top Features :")
print(chi_scores.head(10)['Feature'].tolist())

print("\n4. ANOVA F-Test Top Features :")
print(anova_scores.head(10)['Feature'].tolist())

print("\n5. Mutual Information Top Features :")
print(mi_df.head(10)['Feature'].tolist())
```
